```python
import re
import sys
import matplotlib.pyplot as plt
import os
```

```python
def parse_log(path):
    """
    Parse the scheduler output log and return a list of events:
    (timestamp_ms, pid, task_name).
    """
    print(path)
    pattern = re.compile(r"\((?P<name>[^:]+):(?P<pid>\d+)\)/\d+/(?P<time>\d+)ms")
    events = []
    with open(path, 'r') as f:
        for line in f:
            m = pattern.search(line)
            if m:
                t = int(m.group('time'))
                pid = int(m.group('pid'))
                name = m.group('name')
                events.append((t, pid, name))
    events.sort(key=lambda x: x[0])
    return events

def compute_total_times(events):
    """
    Given a sorted list of (time, pid, name) events, compute how long each
    task ran by subtracting consecutive timestamps.
    """
    totals = {}
    prev = None
    for t, pid, name in events:
        if prev:
            prev_t, prev_pid, prev_name = prev
            dt = t - prev_t
            key = (prev_pid, prev_name)
            totals[key] = totals.get(key, 0) + dt
        prev = (t, pid, name)
    return totals

def plot_totals(totals, log_path):
    """
    Create a bar graph of total run times (ms) per task.
    """
    labels = [f"{name}:{pid}" for (pid, name) in totals.keys()]
    times = list(totals.values())

    if len(totals) > 16:
        # Get the indices of the largest and smallest values
        sorted_indices = sorted(range(len(times)), key=lambda i: times[i])
        smallest_indices = sorted_indices[:8]  # Select the smallest 8 values
        largest_indices = sorted_indices[-8:]  # Select the largest 8 values
        selected_indices = set(smallest_indices + largest_indices)

        # Add spacing between labels
        labels = [labels[i] if i in selected_indices else "" for i in range(len(labels))]
        for i in range(len(labels)):
            if labels[i] != "" and i > 0 and labels[i - 1] != "":
                labels[i] = "\n" + labels[i]  # Add spacing by inserting a newline
        colors = ["red" if i in largest_indices else "blue" if i in smallest_indices else "skyblue" for i in range(len(labels))]
    else:
        colors = ["skyblue"] * len(labels)  # Default color for all bars if <= 16 tasks

    plt.figure(figsize=(8,4))
    plt.bar(range(len(times)), times, color=colors, tick_label=labels)
    plt.ylabel("Average runtime (ms)")
    plt.title("Average CPU Time per Task" + f" (from {log_path})")
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

def print_totals(totals):
    """
    Print total run times per task to stdout.
    """
    print("Total run times per task (ms):")
    for (pid, name), runtime in totals.items():
        print(f"  {name}:{pid} → {runtime} ms")

def compute_average_waits(events):
    """
    For each task, compute the average time between its consecutive CPU‐start events.
    """
    times = {}
    for t, pid, name in events:
        if(name == "Init"):
            continue
        key = (pid, name)
        times.setdefault(key, []).append(t)

    avg_waits = {}
    for key, ts in times.items():
        if len(ts) < 2:
            avg_waits[key] = 0.0
        else:
            intervals = [ts[i] - ts[i-1] for i in range(1, len(ts))]
            avg_waits[key] = sum(intervals) / len(intervals)
    return avg_waits

def plot_avg_waits(avg_waits, log_path):
    """
    Bar‐chart of average wait times per task.
    """
    labels = [f"{name}:{pid}" for (pid, name) in avg_waits.keys()]
    waits  = list(avg_waits.values())

    if len(avg_waits) > 16:
        # Get the indices of the largest and smallest values
        sorted_indices = sorted(range(len(waits)), key=lambda i: waits[i])
        smallest_indices = sorted_indices[:8]  # Select the smallest 8 values
        largest_indices = sorted_indices[-8:]  # Select the largest 8 values
        selected_indices = set(smallest_indices + largest_indices)

        # Add spacing between labels
        labels = [labels[i] if i in selected_indices else "" for i in range(len(labels))]
        for i in range(len(labels)):
            if labels[i] != "" and i > 0 and labels[i - 1] != "":
                labels[i] = "\n" + labels[i]  # Add spacing by inserting a newline
        colors = ["red" if i in largest_indices else "blue" if i in smallest_indices else "salmon" for i in range(len(labels))]
    else:
        colors = ["salmon"] * len(labels)  # Default color for all bars if <= 16 tasks


    plt.figure(figsize=(8,4))
    plt.bar(range(len(waits)), waits, color=colors, tick_label=labels)
    plt.ylabel("Average wait time (ms)")
    plt.title("Average Wait Time per Task" + f" (from {log_path})")
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

def is_one_letter_difference(str1, str2):
    """
    Check if two strings differ by exactly one letter.
    """
    if len(str1) != len(str2):
        return False
    diff_count = sum(1 for a, b in zip(str1, str2) if a != b)
    return diff_count == 1

def average_totals(*totals_list):
    """
    Compute the average value for each key across multiple totals dictionaries.
    
    Parameters:
        *totals_list: A variable number of totals dictionaries to average.
    
    Returns:
        A dictionary with the same keys as the input dictionaries, where each value
        is the average of the corresponding values across all input dictionaries.
    """
    combined_totals = {}
    count = {}

    for totals in totals_list:
        for key, value in totals.items():
            combined_totals[key] = combined_totals.get(key, 0) + value
            count[key] = count.get(key, 0) + 1

    # Compute the average for each key
    averaged_totals = {key: combined_totals[key] / count[key] for key in combined_totals}
    return averaged_totals

def average_waits(*waits_list):
    """
    Compute the average value for each key across multiple wait dictionaries.
    
    Parameters:
        *waits_list: A variable number of wait dictionaries to average.
    
    Returns:
        A dictionary with the same keys as the input dictionaries, where each value
        is the average of the corresponding values across all input dictionaries.
    """
    combined_waits = {}
    count = {}

    for waits in waits_list:
        for key, value in waits.items():
            combined_waits[key] = combined_waits.get(key, 0) + value
            count[key] = count.get(key, 0) + 1

    # Compute the average for each key
    averaged_waits = {key: combined_waits[key] / count[key] for key in combined_waits}
    return averaged_waits

if __name__ == "__main__":
    path = os.path.join(os.getcwd(), "results")
    file_list = [log_file for log_file in os.listdir(path)]
    file_list.sort()
    print(file_list)
    allavg= []
    allTotals = []

    interactive = {0, 2, 4, 6}
    interactiveNoCalc = {1, 3, 5, 7}

    interactiveSpawner = {8, 10 , 12 , 14}
    interactiveSpawnerNoCalc = {9, 11 , 13 , 15}
    
    mixedSpawner = {16, 18, 20, 22}
    mixedSpawnerNoCalc = {17, 19, 21, 23}
    
    NonInteractive = {24, 26, 28, 30}
    NonInteractiveNoCalc = {25, 27, 29, 31}
    
    NonInteractiveSpawner = {32, 34, 36, 38}
    NonInteractiveSpawnerNoCalc = {33, 35, 37, 39}

    profiles = [
            interactive, interactiveNoCalc,
                NonInteractive, NonInteractiveNoCalc,
                interactiveSpawner, interactiveSpawnerNoCalc,
                NonInteractiveSpawner, NonInteractiveSpawnerNoCalc,
                mixedSpawner, mixedSpawnerNoCalc
    ]
    log_Names = [
            "Interactive", "InteractiveNoCalcReadyQueue",
                "NonInteractive", "NonInteractiveNoCalcReadyQueue",
                "InteractiveSpawner", "InteractiveSpawnerNoCalcReadyQueue",
                "NonInteractiveSpawner", "NonInteractiveSpawnerNoCalcReadyQueue",
                "MixedSpawner", "MixedSpawnerNoCalcReadyQueue"
    ]
    for i in range(len(file_list)):
        log_file = file_list[i]

        print(f"Processing log file: {log_file}")
        events = parse_log(path + f"//{log_file}")

        totals = compute_total_times(events)
        allTotals.append(totals)

        avg_waits = compute_average_waits(events)
        allavg.append(avg_waits)

    print(len(allTotals))
    for k in range(len(profiles)):
            average = average_totals(*(allTotals[j] for j in profiles[k]))
            print("\nAverage cpu time per task (ms):")
            print_totals(average)
            plot_totals(average, log_Names[k])
            sumWaits = compute_average_waits(sumEvents)
            avg_waits = average_waits(*(allavg[j] for j in profiles[k]))
            print("\nAverage wait times per task (ms):")
            for (pid, name), w in avg_waits.items():
                print(f"  {name}:{pid} → {w:.2f} ms")
            plot_avg_waits(avg_waits, log_Names[k])
            sumEvents.clear() 
        
```
