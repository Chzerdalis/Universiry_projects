After the installation of the **Anaconda** package (https://www.anaconda.com/products/distribution), you should do the following:

# 1) Install nbextensions

Run **Anaconda Prompt as Administrator** and type the following:

**conda install -c conda-forge jupyter_contrib_nbextensions**

**jupyter contrib nbextension install --user** 

**jupyter nbextension enable varInspector/main**

# 2) Install Latex Environments

Run **Anaconda Prompt as Administrator** and type the following:

**pip3 install jupyter_latex_envs**

**jupyter nbextension install --py latex_envs**

**jupyter nbextension enable --py latex_envs**

Now, restart **Jupyter Notebook**.

From now on, when you launch Jupyter Notebook, in the Home Page, a new tab should appear, named **Nbextensions**.

In this tab, the option "disable configuration for nbextensions" should be deselected and **(some) LaTeX environments for Jupyter** should be enabled.

If everything is done correctly, the following Latex expression should appear normally.

\begin{itemize}
\item item1
\item item2
\item item3
\end{itemize}

# 3) Install R Kernel

After the installation of R (https://cran.r-project.org/) and RStudio (you can download it from the **Anaconda Navigator** or https://www.rstudio.com/products/rstudio/download/), you are now ready to integrate the R kernel into Jupyter Notebook.

Run **Anaconda Prompt as Administrator** and type the following:

**conda install -c r r-irkernel**

Now, **restart** Jupyter Notebook.

In Jupyter Notebook, select **Kernel > Change Kernel > R**

If everything is done correctly, the execution of the following cell will show the R version you are currently using.

```python
R.version
```

# 4) Install Julia Kernel

After the installation of Julia (https://julialang.org/downloads/), you are now ready to integrate the Julia kernel into Jupyter Notebook.

Run **Julia** and in the command line type the following:

**using Pkg**

**Pkg.add("IJulia")**

Now, **restart** Jupyter Notebook.

In Jupyter Notebook, select **Kernel > Change Kernel > Julia**

If everything is done correctly, the execution of the following cell will show the Julia version you are currently using.

```python
VERSION
```

# 5) Installing Matlab

You can download Matlab trial version (https://www.mathworks.com/products/matlab.html) from the official site, by creating a Student account and providing your university email. After the installation is completed, you should run the "Activate Matlab" program, to activate your trial version. Use the activation with internet by logging in your MathWorks account. After that, you are ready to integrate Matlab kernel into Jupyter Notebook.

## Check if your version of python is compatible with your version of Matlab

```python
from IPython.display import Image
Image("matlab-python_compatibility.png")
```

Run **Anaconda Prompt as Administrator** and type the following:

**python --version**

### Important:

*Only if your version of python is compatible with the version of Matlab that you previously downloaded, just run:*

**conda install -c conda-forge matlab_kernel**

*and then just follow the instructions that are listed in **Installing the Matlab engine** chapter.*

**If your version of python is not compatible with the version of Matlab that you previously downloaded, you can then either downgrade your anaconda system's python version, or create a new enviroment that uses a suitable version:**

## Matlab-Python Incompatibility

### Downgrading the 'base' environment

Just type in your Anaconda Prompt:

**conda install python=x.y**

, where 'x.y', the version of python that is compatible with your Matlab version, as listed above.

This might be the simplest solution available, however, *there is a chance of package collision ocurring between your old and new version of python.*

If such a collision occurs, it is preferable to create a new enviroment, rather than downgrading the *base*.

### Creating a new enviroment (only if 'Downgrading the base environment' does not work)

To create a new environment named "jmatlab" (or whichever name you might want to give it), run the following **commands** in the Anaconda Prompt:

**conda create -vv -n jmatlab python=x.y jupyter**, where 'x.y', the version of python that is compatible with your Matlab version, as listed above and "jmatlab" the name that we gave to our new environment, change it accordingly

**conda activate jmatlab**, also, everytime you open the Anaconda Prompt, for you to use the Matlab kernel, you will have to activate this particular environment by typing this command

### Installing the Matlab Kernel

Just simply type in Anaconda Prompt,

**conda install -c conda-forge matlab_kernel**

## Installing the Matlab engine

Now, in order to install the Matlab engine for python, you should find the Matlab installation directory (usually is C:\Program Files\MATLAB). This directory contains a folder named as your Matlab version (e.g., R2021b). In the **Anaconda Prompt** you should type the following command (the path may not be exactly the same, the Matlab version or the installation path may differ).

**cd "C:\Program Files\MATLAB\R2022b\extern\engines\python"**

and now type:

**python setup.py install**

**Restart** Jupyter Notebook.

In Jupyter Notebook, select **Kernel > Change Kernel > Matlab**

If everything is done correctly, the execution of the following cell will show the Matlab version you are currently using.

```python
version
```
