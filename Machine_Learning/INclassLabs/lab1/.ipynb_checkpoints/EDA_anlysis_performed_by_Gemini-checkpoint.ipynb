```python
import pandas as pd
import numpy as np
import dtale
import seaborn as sns
import matplotlib.pyplot as plt
import sweetviz as sv
from ydata_profiling import ProfileReport
```

```python
df = sns.load_dataset('tips')
```

```python
df.head()
df.tail()
df.describe()
df.info()
dtale.show(df)
```

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['total_bill'], bins=20, kde=True)
plt.title('Distribution of Total Bill')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='total_bill', y='tip', data=df, hue='sex')
plt.title('Total Bill vs. Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='day', y='tip', data=df)
plt.title('Tip Distribution by Day')
plt.xlabel('Day')
plt.ylabel('Tip')
plt.show()
```

```python
non_numerical_cols = df.select_dtypes(exclude=['number']).columns
print("Non-numerical columns:", non_numerical_cols)
```

```python
numerical_df = df.select_dtypes(include=['number'])
correlation_matrix = numerical_df.corr()
encoded_df = pd.get_dummies(df, columns=non_numerical_cols)
correlation_matrix = encoded_df.corr()
```

```python
print("Correlation Matrix:\n", correlation_matrix)
```

```python
correlation = df['total_bill'].corr(df['tip'])
print("\nCorrelation between total_bill and tip:", correlation)
```

```python
report = sv.analyze(df)
report.show_html('sweetviz_report.html')
```

```python
profile = ProfileReport(df, title="Pandas Profiling Report", explorative=True)
profile.to_file("pandas_profiling_report.html")
```
