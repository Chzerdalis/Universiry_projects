```python
import pandas as pd
import nltk
nltk.download('punkt_tab')
from nltk.tokenize import sent_tokenize
```

```python
text = """
The alarm rang at 7 in the morning as it usually did on Tuesdays. She rolled over,
stretched her arm, and stumbled to the button till she finally managed to switch it off.
Reluctantly, she got up and went for a shower. The water was cold as the day before the engineers
did not manage to get the boiler working. Good thing it was still summer.
Upstairs, her cat waited eagerly for his morning snack. Miaow! He voiced with excitement
as he saw her climb the stairs.
"""
```

```python
# separate text into sentences

sent_tokenize(text)
```

```python
# count number of sentences

len(sent_tokenize(text))
```

```python
df = pd.read_csv(r'C:\Users\Βαγγέλης\Desktop\Σχολή\4o Έτος\7ο Εξάμηνο\Μηχανική Μάθηση για Επιστήμη Δεδομένων και Ανάλυση\Labs\Lab 5 - Exracting Features from Text\imdb_dataset.csv')
df.rename(columns={'review': 'text'}, inplace=True)

# Example: Tokenize sentences for a sample review
sample_text = df['text'][0]
sentences = sent_tokenize(sample_text)
sentence_count = len(sentences)

# Apply sentence tokenization to the entire dataset
df['num_sentences'] = df['text'].apply(lambda x: len(sent_tokenize(x)))

# Display the updated DataFrame with the new sentence count feature
df[['text', 'num_sentences']].head()
```

```python
# take the first 10 rows to speed things up

df = df.loc[1:10]
df
```

```python
import re

# Remove HTML tags (e.g., <br />) from each review
df['text'] = df['text'].apply(lambda x: re.sub(r'<.*?>', '', x))

# Display the cleaned DataFrame
df.head()
```
