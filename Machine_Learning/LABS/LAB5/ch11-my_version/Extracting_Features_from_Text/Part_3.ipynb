```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

```python
df = pd.read_csv(r'C:\Users\Βαγγέλης\Desktop\Σχολή\4o Έτος\7ο Εξάμηνο\Μηχανική Μάθηση για Επιστήμη Δεδομένων και Ανάλυση\Labs\Lab 5 - Exracting Features from Text\imdb_dataset.csv')
df.rename(columns={'review': 'text'}, inplace=True)

df.head()
```

```python
# Remove punctuation and numbers
df['text'] = df['text'].str.replace(r'[^\w\s]', '', regex=True).str.replace(r'\d+', '', regex=True)

df.head()
```

```python
# set up a bag of words transformer

vectorizer = CountVectorizer(lowercase=True,
                             stop_words='english',
                             ngram_range=(1, 1),
                             min_df=0.05)
```

```python
# transformer finds the words to be retained

vectorizer.fit(df['text'])
```

```python
X = vectorizer.transform(df['text'])
```

```python
# create bago of words dataframe

bagofwords = pd.DataFrame(X.toarray(),
                          columns = vectorizer.get_feature_names_out())

bagofwords.head()
```
