```python
import matplotlib.pyplot as plt

def plot_features(df, text_var):
    # Set up grid dimensions
    nb_rows = 1
    nb_cols = 2

    # Create a figure and set up subplots
    fig, axs = plt.subplots(nb_rows, nb_cols, figsize=(12, 6))
    plt.subplots_adjust(wspace=0.4, hspace=0.4)

    # Plot histograms for each sentiment category
    sentiments = df['sentiment'].unique()
    for i, sentiment in enumerate(sentiments):
        axs[i].hist(df[df['sentiment'] == sentiment][text_var], bins=30)
        axs[i].set_title(f"{text_var} | {sentiment}")
        axs[i].set_xlabel(text_var)
        axs[i].set_ylabel("Frequency")

    plt.show()

# Example call to plot based on word count feature
plot_features(df, 'num_words')
```
