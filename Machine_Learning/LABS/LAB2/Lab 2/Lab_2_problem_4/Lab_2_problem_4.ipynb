```python
import pandas as pd
from sklearn.feature_selection import chi2
from sklearn.feature_selection import SelectKBest
```

```python
file_path = r'C:\Users\Βαγγέλης\Desktop\Σχολή\4o Έτος\7ο Εξάμηνο\Μηχανική Μάθηση για Επιστήμη Δεδομένων και Ανάλυση\In-class Labs\inclass Lab 1 - EDA in python\synthetic_covid19_dataset.csv'
df = pd.read_csv(file_path)
```

tha xrhsimoppoisoume diafores methodoous xehwrista h mia apo thn allh kai oxi diadoxika

```python
#dialegoume to "Decreased Consciousness" ws target variable
#arxika trexontas thn entolh X = df.drop(columns=['Decreased Consciousness']) phrame to output : "Input X must be non-negative" , opote auto pou kaname einai na prosthesoume se ola ta features thn apoluth timh tou min kai na prosthesoume 1 gia na ginoun ola thetika
X = df.drop(columns=['Decreased Consciousness']).apply(lambda x: x + abs(x.min()) + 1 if x.min() < 0 else x)
y = df['Decreased Consciousness']
```

```python
#efarmogh ths methodou chi square test
selector = SelectKBest(chi2, k=4)
X_new = selector.fit_transform(X, y)

selected_features = X.columns[selector.get_support()]

print("Ta features pou dialehtikan xrisimopoiwntas Chi-Square Test:", selected_features)
```

```python
#efarmogh ths methodou forward selection
from sklearn.feature_selection import SequentialFeatureSelector
from sklearn.linear_model import LinearRegression

model = LinearRegression()
sfs = SequentialFeatureSelector(model, n_features_to_select=4, direction='forward')
X_new_1 = sfs.fit_transform(X, y)

selected_features_fs = X.columns[sfs.get_support()]
print("Ta features pou dialehtikan xrisimopoiwntas forward selection method ", selected_features_fs)
```

```python
#efarmogh ths methodou random forest classifier
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
X_new_2 = model.fit(X, y)
importance = model.feature_importances_

selected_features_rf = X.columns[importance.argsort()[-4:][::-1]]
print("Ta features pou dialehtikan xrisimopoiwntas random forest classifier ", selected_features_rf)
```

```python
#efarmogh ths methodou LDA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

lda = LinearDiscriminantAnalysis(n_components=1)
lda.fit(X, y)

coefficients = lda.coef_.flatten()
selected_indices = coefficients.argsort()[-4:][::-1]

selected_features_lda = X.columns[selected_indices]
print("Ta features pou dialehtikan xrisimopoiwntas LDA ", selected_features_lda)
```

```python
#efarmogh hybrid methodou
from sklearn.feature_selection import SelectFromModel
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier()
selector = SelectFromModel(model)
X_new = selector.fit_transform(X, y)

selected_features = X.columns[selector.get_support()]
print("Ta features pou dialehtikan xrisimopoiwntas hybrid method ", selected_features)
```

Vlepoume oti ta features pou dialehtikan xrisimopoiwntas to chi square test kai to forward selection method einai idia. To idio isxuei kai gia ta features pou dialehtikan xrisimopoiwntas to random forest classifier kai to hybrid method. Ta features pou dialehtikan xrisimopoiwntas to LDA einai diaforetika apo ta proigoumena. To gegonos oti yparxoun diafores stis methodous einai logiko, giati kathe methodos exei diaforetiko tropo leitourgias kai diaforetika kritiria gia thn epilogh twn features.
